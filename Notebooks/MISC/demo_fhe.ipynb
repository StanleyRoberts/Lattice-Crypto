We then pick a (single bit) message for alice to encrypt using bobs public key\
Here is the resulting ciphertext:

```python
# remember plaintext messages are represented as polynomial coefficients, up to some plaintext modulus value
message = [20, 11, 2, 12, 4, 3, 1, 5, 1]


bob_public_key = bob.getPublicKey()
cipher = alice.encrypt(message, bob_public_key) # encrypt message using Bob's public key
print(cipher)
```

Bob then decrypts this using his own secret key.\
Here is the plaintext bob gets by decrypting:

```python
plain = bob.decrypt(cipher)

# we take the last 10 values, as the full plaintext is a sparse vector as long as the ciphertext polynomial degree
print(plain[-10:])
```

Fully Homomorphic Functions
-------------------------------------------

Here we will try out the new homomorphic functions: addition and multiplication\
Remember we apply the functions to the ciphertexts but expect the final decrypted result to be the same as applying the operation to the two plaintexts

First lets try multiplication:

```python
key = bob.getPublicKey()

a = alice.encrypt([0, 1, 15, 2, 8, 11, 2], key)
b = alice.encrypt([1, 18, 9, 3, 4, 8, 4], key)

mul_cipher = a*b
print(mul_cipher)
```

\
As you can see we get the same result as we would by just multiplying the two plaintexts!\
(Remember the resulting coefficients are applied some plaintext modulus, in this case the modulus is the default 20)

```python
print(bob.decrypt(mul_cipher)[-12:])
```

<br>
now lets demonstrate addition:

```python
add_cipher = a+b
print(add_cipher)
```

\
decrypting this ciphertext gets us the expected value of addition!

```python
print(bob.decrypt(add_cipher)[-12:])
```
